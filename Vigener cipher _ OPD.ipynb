### Vigener Cipher


```python
def generateKey(key,text):
  key = list(key)
  if len(text) == len(key):
      return(key)
  else:
      for i in range(len(text) - len(key)):
          key.append(key[i % len(key)])
  return ("" . join(key))

def encrypt(text,key):
  result = []
  for i in range(len(text)):
      x = (ord(text[i]) + ord(key[i])) % 26
      x += ord('A')
      result.append(chr(x))
  return ("" . join(result))
  
def decrypt(text,key):
  result = []
  for i in range(len(text)):
      x = (ord(text[i]) - ord(key[i]) + 26) % 26
      x += ord('A')
      result.append(chr(x))
  return ("" . join(result))

text = input("Enter the plain text: ")
key = input("Enter the key: ")

generatedKey = generateKey(key,text)

encrypted = encrypt(text,generatedKey)
print("Encrypted Text: "+encrypted)

decrypted = decrypt(encrypted,generatedKey)
print("Decrypted Text: "+decrypted)
```

### One Time Pad Cipher

```python
def otpEncrypt(text,key):
  result = []
  for i in range(len(key)):
    if text[i].isalpha():
      result.append(ord(text[i]) - ord('A') + ord(key[i]) -ord('A'))
    else:
      result.append(text[i])

  for i in range(len(result)):
    if result[i] != " ":
      if result[i] > 25:
        result[i] = result[i] -26

  res = ""

  for i in result:
    if i != " ":
      res += chr(i + ord('A'))
    else:
      res += i 
  return (res)

def otpDecrypt(text,key):
  result = []
  for i in range(len(key)):
    if text[i].isalpha():
      result.append(ord(text[i]) - ord('A') - (ord(key[i]) -ord('A')))
    else:
      result.append(text[i])

  for i in range(len(result)):
    if result[i] != " ":
      if result[i] < 0:
        result[i] = result[i] +26

  res = ""

  for i in result:
    if i != " ":
      res += chr(i + ord('A'))
    else:
      res += i 
  return (res)

text = input("Enter the plain text: ")
key = input("Enter the key: ")
encrypted = otpEncrypt(text,key)
print("Encrypted text: "+encrypted)
decrypted = otpDecrypt(encrypted,key)
print("Decrypted text: "+decrypted)
```
